```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def spring_mass_dashpot(t, y, m, c, k):
    x, v = y
    dxdt = v
    dvdt = -(c/m)*v - (k/m)*x + np.sin(2*np.pi*t)
    return dxdt, dvdt

t_eval = np.linspace(0, 20/2/np.pi, 1000)
sol = solve_ivp(lambda t, y: spring_mass_dashpot(t, y, 1, 2.5, 60), [0, 20/2/np.pi], [1.0, 5.0], t_eval=t_eval)

plt.plot(sol.t, sol.y[0])
plt.xlabel('Time (s)')
plt.ylabel('Displacement (cm)')
plt.grid()
plt.show()


peak = max(sol.y[0]) - min(sol.y[0])
peak_positive = peak / 2
print("Peak positive amplitude: ", peak_positive, " cm")
```
